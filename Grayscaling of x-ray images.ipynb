```python
# # Grey scaling of images for train data
import cv2

import os
```

```python
# Set the path to the folder containing the images

folder_path = '/content/drive/MyDrive/Colab_Notebooks/images/'
```

```python
# Create a new folder to save the grayscale images

output_folder_path = '/content/drive/MyDrive/Colab_Notebooks/gray_images/'

os.makedirs(output_folder_path, exist_ok=True)
```

```python
# Loop through each file in the folder

for file_name in os.listdir(folder_path):

    # Load the image
    file_path = os.path.join(folder_path, file_name)
    img = cv2.imread(file_path)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Write the grayscale image to the output folder
    output_file_path = os.path.join(output_folder_path, file_name)
    cv2.imwrite(output_file_path, gray)
```
