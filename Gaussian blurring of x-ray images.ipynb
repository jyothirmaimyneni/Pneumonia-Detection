```python
# Gaussian Blur

import os
import cv2
```

```python
# Define the paths of the input and output folders

input_folder_path = '/content/drive/MyDrive/Colab_Notebooks/images_pnemonia_train/'
output_folder_path = '/content/drive/MyDrive/Colab_Notebooks/gaussianblur_pnemonia_train'
```

```python
# Create the output folder if it doesn't exist

if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)
```

```python
# Loop through all the files in the input folder

for filename in os.listdir(input_folder_path):

    # Load the image
    img_path = os.path.join(input_folder_path, filename)
    img = cv2.imread(img_path)

    # Apply Gaussian blur
    img_blur = cv2.GaussianBlur(img, (5, 5), 0)

   # Save the blurred image to the output folder
    output_path = os.path.join(output_folder_path, filename)
    cv2.imwrite(output_path, img_blur)
```
