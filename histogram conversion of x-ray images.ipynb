```python
#Changing X-ray Image to Histogram

import cv2

import matplotlib.pyplot as plt

import os


```

```python
# Define the paths to the input and output folders

input_folder = '/content/drive/MyDrive/Colab_Notebooks/images_pnemonia_test/'
output_folder = '/content/drive/MyDrive/Colab_Notebooks/histograms_pnemonia_test'
```

```python
# Create the output folder if it does not exist

if not os.path.exists(output_folder):

    os.makedirs(output_folder)

```

```python
# Loop through all the files in the input folder

for filename in os.listdir(input_folder):

    # Load the image
    img = cv2.imread(os.path.join(input_folder, filename), cv2.IMREAD_GRAYSCALE)

    # Calculate the histogram
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])

    # Plot the histogram

    plt.plot(hist)
    plt.xlim([0, 256])
    plt.title('Histogram of ' + filename)
    plt.xlabel('Pixel intensity')
    plt.ylabel('Frequency')

    # Save the histogram to the output folder
    output_path = os.path.join(output_folder, os.path.splitext(filename)[0] + '_histogram.jpg')
    plt.savefig(output_path)

    # Clear the plot for the next image

    plt.clf()
```
